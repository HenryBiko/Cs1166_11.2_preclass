```python
%matplotlib inline
import seaborn as sns 
sns.set()
from scipy.stats import norm
from random import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    Z = 24.44321494051954
    if abs(x) > 7:
        return 0
    elif abs(x) > 3:
        return 3 * (1 - (x / 7) ** 2) ** 0.5 / Z
    elif abs(x) > 1:
        return (
            (3 - abs(x)) / 2 -
            3/7 * 10**0.5 * ((3 - x**2 + 2*abs(x))**0.5 - 2)
        ) / Z
    elif abs(x) > 0.75:
        return (9 - 8 * abs(x)) / Z
    elif abs(x) > 0.5:
        return (3 * abs(x) + 0.75) / Z
    else:
        return 2.25 / Z
```

```python
s = [0]
for i in range(100000):
    proposal =function(s[-1])
    p = min(1, f(proposal) / f(s[-1]))
    if np.random.uniform(0, 1) < p:
        s.append(proposal)
    else:
        s.append(s[-1])      
def function(x):
    return np.random.normal(x, 2)
```

```python

plt.hist(s[100::100],density=True, bins=30)
x = np.linspace(-8, 8, 561)
y = [f(i) for i in x]
plt.plot(x, y, 'r-', color="red")
```
